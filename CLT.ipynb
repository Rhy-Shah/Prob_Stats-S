Systolic blood pressure of a group of people is known to have an average of 122 mmHg and a standard deviation of 10 mmHg Calculate the probability that the average blood pressure of 16 people will be greater than 125 mmHg.

```python
1 - norm.cdf(1.2)
```

```python
1 - norm.cdf(x=125,loc=122,scale=2.5)
```

Weekly toothpaste sales have a mean 1000 and std dev 200. What is the probability that the average weekly sales next month is more than 1110?

```python
1 - norm.cdf(x=1110,loc=1000,scale=100)
```

In an ecommerce website, the average purchase amount per bill is $80 with a standard deviation of $15. If we randomly select a sample of 50 customers, what is the probabiltity that the average purchase amount in the sample will be less than $75

```python
norm.cdf((75-80)/(15/np.sqrt(50)))
```

```python
norm.cdf(x=75,loc=80,scale=(15/np.sqrt(50)))
```
